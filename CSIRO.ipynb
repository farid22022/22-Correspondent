# Task
Analyze the provided data, build a model to predict pasture biomass based on the given features and image data, and generate a submission file in the specified format.

## Explore and preprocess data

### Subtask:
Analyze the structure and content of the training data, including image paths, sampling dates, state, species, NDVI, height, target names, and target values. Handle missing values and convert the 'Sampling_Date' column to datetime objects.


**Reasoning**:
Display the column names and their data types of the `df_train` DataFrame and check for missing values.



```python
display(df_train.info())
display(df_train.isnull().sum())
```

**Reasoning**:
Convert the 'Sampling_Date' column to datetime objects and display the first few rows to verify the conversion.



```python
df_train['Sampling_Date'] = pd.to_datetime(df_train['Sampling_Date'])
display(df_train.head())
```

## Feature engineering

### Subtask:
Extract relevant features from the image paths (e.g., image IDs), sampling dates (e.g., year, month, day), and other columns. Consider creating interaction terms or polynomial features if necessary.


**Reasoning**:
Extract image ID, year, month, day, and day of year from the relevant columns and display the updated dataframe.



```python
df_train['image_id'] = df_train['image_path'].apply(lambda x: x.split('/')[-1].split('.')[0])
df_train['year'] = df_train['Sampling_Date'].dt.year
df_train['month'] = df_train['Sampling_Date'].dt.month
df_train['day'] = df_train['Sampling_Date'].dt.day
df_train['day_of_year'] = df_train['Sampling_Date'].dt.dayofyear
display(df_train.head())
```

## Image data processing

### Subtask:
Determine how to incorporate image data into the model. This could involve using pre-trained convolutional neural networks (CNNs) to extract features or resizing and flattening images for use with traditional machine learning models.


## Model selection and training

### Subtask:
Choose appropriate regression models (e.g., linear regression, decision trees, random forests, gradient boosting models, neural networks) for each target variable. Train the selected models using the preprocessed data and engineered features.


**Reasoning**:
Identify the unique target variables, filter the training data for each target, define features and target, choose and train a regression model (using RandomForestRegressor as a starting point), and store the trained model in a dictionary.



```python
from sklearn.ensemble import RandomForestRegressor

unique_targets = df_train['target_name'].unique()
trained_models = {}

for target_name in unique_targets:
    df_subset = df_train[df_train['target_name'] == target_name].copy()


    X = df_subset.select_dtypes(include=np.number).drop(columns=['target'])
    y = df_subset['target']


    model = RandomForestRegressor(random_state=42)
    model.fit(X, y)


    trained_models[target_name] = model

print("Trained models for each target variable:")
for target_name, model in trained_models.items():
    print(f"- {target_name}: {type(model).__name__}")
```

## Prediction

### Subtask:
Use the trained models to predict biomass values for the test set.


**Reasoning**:
Prepare the test data by extracting date features and then iterate through the trained models to make predictions for each target name in the test set, storing the results.



```python
df_test['Sampling_Date'] = pd.to_datetime('2015-01-01') # Assign a dummy date as test set does not have it.
df_test['image_id'] = df_test['image_path'].apply(lambda x: x.split('/')[-1].split('.')[0])
df_test['year'] = df_test['Sampling_Date'].dt.year
df_test['month'] = df_test['Sampling_Date'].dt.month
df_test['day'] = df_test['Sampling_Date'].dt.day
df_test['day_of_year'] = df_test['Sampling_Date'].dt.dayofyear

predictions = []

for target_name, model in trained_models.items():
    df_test_subset = df_test[df_test['target_name'] == target_name].copy()


    X_test = df_test_subset.select_dtypes(include=np.number)


    y_pred = model.predict(X_test)


    for i, sample_id in enumerate(df_test_subset['sample_id']):
        predictions.append({'sample_id': sample_id, 'target': y_pred[i]})

predictions_df = pd.DataFrame(predictions)
display(predictions_df.head())
```

## Evaluation

### Subtask:
Evaluate the model performance using the provided weighted R² metric and iterate on the model selection, feature engineering, and training steps to improve results.


**Reasoning**:
Define the weighted R² metric function and calculate the weighted R² score for the training data predictions.



```python
from sklearn.metrics import r2_score

def weighted_r2_score(y_true, y_pred, weights):

    weighted_mean_true = np.average(y_true, weights=weights)


    weighted_total_ss = np.sum(weights * (y_true - weighted_mean_true)**2)


    weighted_residual_ss = np.sum(weights * (y_true - y_pred)**2)


    weighted_r2 = 1 - (weighted_residual_ss / weighted_total_ss)

    return weighted_r2


y_true_train = df_train['target']

y_pred_train = []
for index, row in df_train.iterrows():
    target_name = row['target_name']
    model = trained_models[target_name]

    X_train_row = row[['year', 'month', 'day', 'day_of_year']].values.reshape(1, -1)
    y_pred_train.append(model.predict(X_train_row)[0])

y_pred_train = np.array(y_pred_train)


weights_train = df_train['target']


weights_train = weights_train + 1e-6


weighted_r2 = weighted_r2_score(y_true_train, y_pred_train, weights_train)

print(f"Weighted R² score on the training data: {weighted_r2}")
```
